```python
# for column in df.columns:
#     plt.figure()
#     sns.histplot(df[column],kde=True)
#     plt.title(f'Distribution of {column}')
#     plt.show()

for column in df.columns:
    sns.histplot(df[column],kde=True)
plt.show()
```

```python
for column in df.columns:
    plt.figure()
    sns.histplot(df[column],kde=True)
    plt.title(f'Distribution of {column}')
    plt.show()
```

```python
sns.pairplot(df)
plt.show()
```

```python
for column in df.columns:
    plt.figure()
    sns.scatterplot(x=df[column],y=df['quality'],hue='quality',data=df)
    plt.title(f'{column} vs quality')
    plt.show()
```

```python
#catplot

sns.catplot(x='quality',y='alcohol',data=df,kind='box')
```
